```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/Kokila412/ML_internship/refs/heads/main/advertising.csv')
df
```

```python
x=df.drop('Sales',axis=1)
x
y=df['Sales']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=model.predict(x_test)
score=model.score(x_test,y_test)
score
```

```python
x_pred=model.predict(x_train)
score=model.score(x_train,y_train)
score
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse_input=mean_squared_error(y_test,y_pred)
r2_input=r2_score(y_test,y_pred)
mse_output=mean_squared_error(y_train,x_pred)
r2_output=r2_score(y_train,x_pred)
dataframe=pd.DataFrame({'mse_input':[mse_input],'r2_input':[r2_input],'mse_output':[mse_output],'r2_output':[r2_output]})
dataframe
```

```python
from matplotlib import pyplot as plt
plt.scatter(y_test,y_pred,alpha=0.5,color='blue')
plt.scatter(y_train,x_pred,alpha=0.5,color='maroon')
plt.xlabel('actual value')
plt.ylabel('predicted value')
plt.title('actual value vs predicted value')
plt.show()
```
